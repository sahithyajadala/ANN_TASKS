```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)

# Compute Softmax function and its derivative
exp_scores = np.exp(x - np.max(x, axis=-1, keepdims=True))
softmax = exp_scores / np.sum(exp_scores, axis=-1, keepdims=True)
softmax_deriv = softmax * (1 - softmax)

# Plot Softmax function and its derivative
plt.figure(figsize=(10, 6))

# Softmax function plot
plt.subplot(2, 1, 1)
plt.plot(x, softmax)
plt.title('Softmax Function')
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.legend(['softmax'])

# Softmax derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, softmax_deriv)
plt.title('Softmax Derivative')
plt.xlabel('x')
plt.ylabel('softmax\'(x)')
plt.legend(['softmax derivative'])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)

# Compute tanh function and its derivative
tanh = np.tanh(x)
tanh_deriv = 1 - np.tanh(x)**2

# Plot tanh function and its derivative
plt.figure(figsize=(10, 6))

# Tanh function plot
plt.subplot(2, 1, 1)
plt.plot(x, tanh, label='tanh', color='blue')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.legend()

# Tanh derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, tanh_deriv, label='tanh derivative', color='green')
plt.title('Tanh Derivative')
plt.xlabel('x')
plt.ylabel('tanh\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)

# Compute ReLU function and its derivative
relu = np.maximum(0, x)
relu_deriv = np.where(x > 0, 1, 0)

# Plot ReLU function and its derivative
plt.figure(figsize=(10, 6))

# ReLU function plot
plt.subplot(2, 1, 1)
plt.plot(x, relu, label='ReLU', color='blue')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()

# ReLU derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, relu_deriv, label='ReLU derivative', color='green')
plt.title('ReLU Derivative')
plt.xlabel('x')
plt.ylabel('ReLU\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)

# Define the alpha parameter for Leaky ReLU
alpha = 0.01

# Compute Leaky ReLU function and its derivative
leaky_relu = np.where(x > 0, x, alpha * x)
leaky_relu_deriv = np.where(x > 0, 1, alpha)

# Plot Leaky ReLU function and its derivative
plt.figure(figsize=(10, 6))

# Leaky ReLU function plot
plt.subplot(2, 1, 1)
plt.plot(x, leaky_relu, label=f'Leaky ReLU (alpha={alpha})', color='blue')
plt.title(f'Leaky ReLU Function (alpha={alpha})')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.legend()

# Leaky ReLU derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, leaky_relu_deriv, label=f'Leaky ReLU derivative (alpha={alpha})', color='green')
plt.title(f'Leaky ReLU Derivative (alpha={alpha})')
plt.xlabel('x')
plt.ylabel('Leaky ReLU\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)

# Compute sigmoid function and its derivative
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_deriv = sigmoid * (1 - sigmoid)

# Plot sigmoid function and its derivative
plt.figure(figsize=(10, 6))

# Sigmoid function plot
plt.subplot(2, 1, 1)
plt.plot(x, sigmoid, label='sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.legend()

# Sigmoid derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, sigmoid_deriv, label='sigmoid derivative', color='green')
plt.title('Sigmoid Derivative')
plt.xlabel('x')
plt.ylabel('sigmoid\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
